```python
from groups.models import Group

group = Group.objects(name="Seed comments - 3 votes")[0]
```

```python
len(group.articles)
```

```python
import json

blacklisted = [
    "\"planero\"",
]

group_articles = group.articles
left_articles = []

for article in group_articles:
    votes = eval(article.description)
    
    sum_without = sum(v for k, v in votes.items() if k not in blacklisted)
    
    if sum_without < 3:
        left_articles.append(article)
```

```python
len(left_articles)
```

ArtÃ­culos que perdemos si sacamos "planero"

```python
for i, article in enumerate(left_articles):
    print(f"{i:<3} -- {article.title}")
```
