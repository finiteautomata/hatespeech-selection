## Negro

Nuestra query actual es

"negro -\"en negro\" -\"Rio\" -\"peste\" -\"humor negro\"",

```python
original_query = "negro -\"en negro\" -\"Rio\" -\"peste\" -\"humor negro\""

Reply.objects.no_dereference().search_text(
    original_query
).count()
```

Hagamos otras queries

```python
queries = [
    "\"negra\" -guita -plata -lagrima -peste -alma -mamba",
    "\"negro de\"",
    "\"negros\"",
]
```

```python
import random

total_results = 0

random.shuffle(queries)

for query in queries:
    for _ in range(3):
        print("*"*80)
    print("\n", "Query: ", query)
    
    results = list(Reply.objects.no_dereference().search_text(query))
    
    total_results += len(results)
    print(len(results), "\n"*2)
    random.shuffle(results)

    results = results[:30]

    for repl in results:
        print("="*80)
        print(repl.text)
```

Veamos la diferencia entre la query original y estas varias

```python
total_results - Reply.objects.no_dereference().search_text(original_query).count()
```
